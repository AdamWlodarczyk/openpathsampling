```python
%matplotlib inline
from toy_plot_helpers import ToyPlot
plot = ToyPlot()
```

```python
import openpathsampling as paths
import numpy as np
```

```python
import openpathsampling.toy_dynamics as toys
pes = (toys.toy_pes.OuterWalls([1.0, 1.0], [0.0, 0.0]) + 
       toys.toy_pes.Gaussian(2.0, [1.0, 4.0], [0.0,1.0]) + 
       toys.toy_pes.Gaussian(2.0, [25.0, 1.0], [0.0, -1.0])
      )
topology=paths.ToyTopology(
    n_spatial = 2,
    masses =[1.0, 1.0],
    pes = pes
)
template_snap = paths.Snapshot(
    coordinates=np.array([[0.0, 0.0]]), 
    velocities=np.array([[0.0,0.0]]),
    potential_energy = 0.0, # to set dimensionless energies for toy engine
    kinetic_energy = 0.0,
    topology=topology
)
integ = toys.toy_integrators.LangevinBAOABIntegrator(dt=0.02, temperature=0.1, gamma=1.0)
options={'integ' : integ,
         'n_frames_max' : 5000,
         'nsteps_per_frame' : 10
        }
sim = toys.toy_engine.ToyEngine(
        options=options,
        template=template_snap
)

paths.PathMover.engine = sim
sim.initialized = True
```

```python
sim.current_snapshot = paths.Snapshot(
    coordinates=np.array([[-0.5, 0.0]]), 
    velocities=np.array([[0.0,0.0]]),
    topology=topology
)
```

```python
plot.add_pes(pes)
plot.add_initial_condition(sim.current_snapshot)
```

```python
plot.plot()
```

```python
orderparameter = paths.OP_Function(name="xval",
                                   fcn=lambda snap : snap.coordinates[0,0])
```

```python
stateA = paths.LambdaVolume(orderparameter, float("-inf"), -0.3)
stateB = paths.LambdaVolume(orderparameter, 0.3, float("inf"))
interfaces = paths.VolumeFactory.LambdaVolumeSet(orderparameter, 
                                                 minvals=float("-inf"), 
                                                 maxvals=[-0.30, -0.25, -0.20, -0.13, -0.05, 0.0]
                                                )
```

```python
plot.add_states([stateA, stateB])
plot.add_interfaces(interfaces)
```

```python
plot.plot()
```

```python
tis = paths.analysis.TISTransition(stateA, stateB, orderparameter, interfaces)
```

```python
tis.ensembles
```

```python
tis.movers
```

```python
first_traj_ensemble = paths.SequentialEnsemble([
    paths.InXEnsemble(stateA),
    paths.OutXEnsemble(stateA),
    paths.InXEnsemble(stateA) & paths.LengthEnsemble(1)
])
```

```python
first_traj = sim.generate(sim.current_snapshot, [first_traj_ensemble.can_append])
```

```python
subtraj = tis.ensembles[0].split(first_traj)[0]
```

```python
plot.plot(bold=[first_traj], trajectories=[subtraj])
```
